```python
import pandas as pd
```

```python
train_df = pd.read_csv("train_df.csv")
test_df = pd.read_csv("test_df.csv")
```

```python
corr_matrix = train_df.corr()
print(corr_matrix)
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
sn.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
train_df["feature_73"].unique()
```

```python
train_df["feature_74"].unique()
```

```python
feature = "feature_"
f_to_drop = []
for i in range(79):
  feature = "feature_" + str(i)
  if len(train_df[feature].unique()) == 1:
    f_to_drop.append(feature)
f_to_drop
```

```python
df = train_df.drop("search_id", axis=1)
df = df.drop(f_to_drop, axis=1)
```

```python
df
```

```python
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
```

```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=10)
pca.fit(X)
X = pca.transform(X)
X
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X, y)
```

```python
clf.decision_path(X)
```

```python
test_df = test_df.drop("search_id", axis=1)
test_df = test_df.drop(f_to_drop, axis=1)
X_test = test_df.iloc[:, 0:-1]
y_test = test_df.iloc[:, -1]
```

```python
X_test = pca.transform(X_test)
```

```python
X_test
```

accuracy

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import ndcg_score

ndcg_score(y_test, y_pred)
```
